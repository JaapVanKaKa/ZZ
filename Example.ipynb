# Exploratory Data Analysis and Visualisation

```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
%matplotlib inline
import time
import random
```

```python
# Importing the dataset
wine = pd.read_csv("https://raw.githubusercontent.com/aniruddhachoudhury/Red-Wine-Quality/master/winequality-red.csv")
```

```python
wine.corr()
```

```python
##  Generates histograms of the numerical columns

# A new Matplotlib figure is created with a size of 15 inches by 12 inches using the figsize parameter.
fig = plt.figure(figsize=(15, 12))

# A title is added to the entire figure using the suptitle function for the entire figure.
plt.suptitle('Histograms of Numerical Columns', fontsize=20)

# A loop is used to iterate over each column in the wine DataFrame.
for i in range(wine.shape[1]):
    # Create a new subplot for the histogram
    plt.subplot(6, 3, i + 1) 
    f = plt.gca()
    # Set the title for the subplot
    f.set_title(wine.columns.values[i]) 
    # Calculate the number of unique values in the column
    # If there are more than 100 unique values, cap the number of bins at 100 
    vals = np.size(wine.iloc[:, i].unique())
    if vals >= 100: 
        vals = 100
    # Create a histogram of the column with the specified number of bins
    plt.hist(wine.iloc[:, i], bins=vals, color='#3F5D7D') 

# Adjust the spacing between the subplots
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
```

```python
# Load the data into a Pandas DataFrame using read_csv() method.

import pandas as pd

df = pd.read_csv("http://www.ishelp.info/data/insurance.csv")
# Explore the data using various methods such as head(), info(), and describe() to get an idea about the data.

# Check for missing values using isnull() and sum() methods.
print(df.isnull().sum())
# Perform some data visualization using libraries like Matplotlib and Seaborn to understand the relationship between variables.

import seaborn as sns
import matplotlib.pyplot as plt

sns.pairplot(df)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Calculate the correlation coefficients
corr = wine.corrwith(wine["quality"]).sort_values(ascending=False)

# Set the color palette for the plot
colors = ["#2ECC71" if c >= 0 else "#E74C3C" for c in corr]

# Plot the correlation coefficients as a bar chart
plt.figure(figsize=(15,8))
plt.bar(corr.index, corr.values, color=colors)
plt.title("Correlation with Wine Quality")
plt.xlabel("Features")
plt.ylabel("Correlation Coefficient")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--')
plt.show()
```

```python
# Generate colors
colors = sns.color_palette('hls', len(wine.corrwith(wine.quality)))
#  Plot the correlation coefficients of all columns with the "quality" column
wine.corrwith(wine.quality).plot.bar(
    figsize=(15, 8), title="Correlation with quality", fontsize=10,
    rot=45, grid=True, color=colors)
```

# Mask Array, Half Heatmap(to hide cells that are not interests) for correlations

```python
# This code generates a heatmap plot of the correlation matrix computed earlier using the corr() function. 
# A mask array is created which is used to hide the upper triangle of the heatmap.
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(18, 15))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(220, 10, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```
